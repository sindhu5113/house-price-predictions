```python
import fitz  # PyMuPDF
import pytesseract
from PIL import Image
import io

# Extract text from scanned images using OCR
def extract_text_with_ocr(pdf_path):
    doc = fitz.open(pdf_path)
    text = ""
    for i in range(len(doc)):
        for img in doc[i].get_images(full=True):
            xref = img[0]
            base_image = doc.extract_image(xref)
            image_bytes = base_image["image"]
            img = Image.open(io.BytesIO(image_bytes))
            text += pytesseract.image_to_string(img) + "\n"
    return text

pdf_path = "/content/Carroll_SG.pdf"
ocr_text = extract_text_with_ocr(pdf_path)
print(f"OCR extracted text length: {len(ocr_text)} characters")

# Save OCR results to file
with open("/content/Carroll_SG.txt.txt", "w", encoding="utf-8") as f:
    f.write(ocr_text)
print("OCR text extraction complete.")
import fitz

def extract_text_from_pdf_fitz(pdf_path):
    doc = fitz.open(pdf_path)
    text = ""
    for page_num in range(len(doc)):
        page_text = doc[page_num].get_text()
        if page_text.strip():
            text += f"--- Page {page_num+1} ---\n" + page_text + "\n"
        else:
            print(f"⚠️ No text found on page {page_num+1}")
    return text

pdf_text_fitz = extract_text_from_pdf_fitz(pdf_path)
print(f"Length of text extracted using fitz: {len(pdf_text_fitz)} characters")

# Save to file to check results manually
with open("/mnt/data/extracted_text_fitz.txt", "w", encoding="utf-8") as f:
    f.write(pdf_text_fitz)
print("Text extraction with PyMuPDF complete.")
```

```python
# Install pytesseract if not already installed
!pip install pytesseract
!sudo apt install tesseract-ocr
import fitz  # PyMuPDF for PDF text extraction
import pytesseract
from PIL import Image
import io
import sympy as sp
from langchain.text_splitter import RecursiveCharacterTextSplitter
from langchain.embeddings import OpenAIEmbeddings
from langchain.vectorstores import FAISS
from langchain.llms import OpenAI
from langchain.chains import RetrievalQA

# Extract text from PDF
def extract_text_from_pdf(pdf_path):
    doc = fitz.open(pdf_path)
    text = ""
    for page in doc:
        text += page.get_text("text")
    return text

# Extract images from PDF
def extract_images_from_pdf(pdf_path):
    doc = fitz.open(pdf_path)
    images = []
    for i in range(len(doc)):
        for img in doc[i].get_images(full=True):
            xref = img[0]
            base_image = doc.extract_image(xref)
            image_bytes = base_image["image"]
            img = Image.open(io.BytesIO(image_bytes))
            images.append(img)
    return images

# Perform OCR on an image
def ocr_image(image):
    return pytesseract.image_to_string(image)

# Convert mathematical expression to SymPy and explain it
def explain_equation(equation_str):
    try:
        expr = sp.sympify(equation_str)
        return sp.pretty(expr)
    except Exception as e:
        return str(e)

# Load and process PDF
pdf_text = extract_text_from_pdf("/content/Carroll_SG.pdf")
images = extract_images_from_pdf("/content/Carroll_SG.pdf")

# Chunking text for retrieval
text_splitter = RecursiveCharacterTextSplitter(chunk_size=500, chunk_overlap=50)
text_chunks = text_splitter.split_text(pdf_text)

# Embeddings and Vector Database
embeddings = OpenAIEmbeddings()
vector_db = FAISS.from_texts(text_chunks, embeddings)

# Load LLM and Retrieval-Augmented Generation (RAG) pipeline
llm = OpenAI(model_name="gpt-3.5-turbo")
rag_chain = RetrievalQA.from_chain_type(
    llm=llm,
    retriever=vector_db.as_retriever(),
    return_source_documents=True
)

def ask_rag(query):
    response = rag_chain({"query": query})
    return response["result"]

# Example Usage
query_result = ask_rag("What is Carroll's theorem about?")
ocr_text = ocr_image(images[0]) if images else "No images found"
equation_explanation = explain_equation("x**2 + 3*x - 5")

# Output Results
print("Query Result:\n", query_result)
print("OCR Extracted Text:\n", ocr_text)
print("Equation Explanation:\n", equation_explanation)
```

```python
# Import required libraries
import fitz  # PyMuPDF for PDF processing
import pytesseract  # OCR for extracting text from images
from PIL import Image  # For working with image data
import io
import sympy as sp  # For handling and explaining mathematical equations

# ---------- Extract Text from PDF ----------
def extract_text_from_pdf(pdf_path):
    doc = fitz.open(pdf_path)
    text = ""
    for page_num in range(len(doc)):
        page_text = doc[page_num].get_text()
        if page_text.strip():  # Check if page contains text
            text += f"--- Page {page_num+1} ---\n" + page_text + "\n"
        else:
            print(f"⚠️ No text found on page {page_num+1}")
    return text

# ---------- Extract Images and Perform OCR ----------
def extract_images_and_ocr(pdf_path):
    doc = fitz.open(pdf_path)
    ocr_text = ""
    for i in range(len(doc)):
        for img_index, img in enumerate(doc[i].get_images(full=True)):
            xref = img[0]
            base_image = doc.extract_image(xref)
            image_bytes = base_image["image"]
            img = Image.open(io.BytesIO(image_bytes))

            # Perform OCR on the image
            extracted_text = pytesseract.image_to_string(img)
            ocr_text += f"--- OCR Text from Page {i+1}, Image {img_index+1} ---\n{extracted_text}\n"

            # Optional: Save the image for inspection
            # The img.save line should be inside the inner loop
            img.save(f"/content/page_{i+1}_img_{img_index+1}.png") # changed directory to '/content'
    return ocr_text

# ---------- Extract and Explain Mathematical Equations ----------
def explain_equation(equation_str):
    try:
        expr = sp.sympify(equation_str)
        explanation = sp.pretty(expr)
        return explanation
    except Exception as e:
        return f"❗ Unable to parse equation: {e}"

# ---------- Manual Chunking of Text ----------
def chunk_text(text, chunk_size=500, chunk_overlap=50):
    chunks = []
    start = 0
    while start < len(text):
        end = start + chunk_size
        chunk = text[start:end]
        chunks.append(chunk)
        start += chunk_size - chunk_overlap
    print(f"✅ Total chunks created: {len(chunks)}")
    return chunks

# ---------- Load PDF and Process Content ----------
pdf_path = "/content/Carroll_SG.pdf"

# Extract text from PDF
pdf_text = extract_text_from_pdf(pdf_path)
print(f"✅ Total length of extracted text: {len(pdf_text)} characters")

# Save extracted text to file
with open("/content/Carroll_SG.txt.txt", "w", encoding="utf-8") as f:
    f.write(pdf_text)

# Extract images and apply OCR
ocr_text = extract_images_and_ocr(pdf_path)
print(f"✅ Total length of OCR extracted text: {len(ocr_text)} characters")

# Save OCR text to file
with open("/content/Carroll_SG.txt.txt", "w", encoding="utf-8") as f:
    f.write(ocr_text)

# ---------- Chunking the Text Manually ----------
text_chunks = chunk_text(pdf_text)

# ---------- Example Usage: Equation Explanation ----------
# Example equation to explain
equation_str = "x**2 + 3*x - 5"
equation_explanation = explain_equation(equation_str)
print("🔢 Equation Explanation:\n", equation_explanation)

# Save all results to a final file
with open("/content/Carroll_SG.txt.txt", "w", encoding="utf-8") as f:
    f.write(f"--- Extracted Text ---\n{pdf_text}\n")
    f.write(f"--- OCR Extracted Text ---\n{ocr_text}\n")
    f.write(f"--- Equation Explanation ---\n{equation_explanation}\n")

print("🎉 Processing complete! Check final_output.txt for results.")
```
